**PARTE I**

```python
import requests as r
```

```python
# Capturando API
url = 'https://api.covid19api.com/dayone/country/brazil'
resp = r.get(url)
```

```python
resp.status_code
```

```python
raw_data = resp.json()
```

```python
# Vendo as variaveis e valores que temos na primeira linha
raw_data[0]
```

```python
# Separando somente as vari√°veis que desejamos
final_data = []
for i in raw_data:
    final_data.append([i['Confirmed'], i['Deaths'], i['Recovered'], i['Active'], i['Date']])
```

```python
final_data
```

```python
# Criando cabe√ßalho:
final_data.insert(0, ['confirmados', 'mortes', 'recuperados', 'ativos', 'data'])
```

```python
final_data
```

```python
# Formatando melhor a data para podermos usar:
for i in range(1, len(final_data)):
    final_data[i][4] = final_data[i][4][:10]              
```

**PARTE II**

```python
def get_datasets(y,labels):
    if type(y[0]) == list:
        datasets = []
        for i in range(len(y)):
            datasets.append({
                'label': labels[i],
                'data': y[i]
            })
        return datasets
    else:
        return [
            {
                'label': labels[0],
                'data': y
            }
        ]
```

```python
def set_title(title=''):
    if title != '':
        display = 'true'
    else:
        display = 'false'
    return {
        'title': title,
        'display': display
    }
```

```python
def create_chart(x, y, labels, kind='bar', title=''):
    
    datasets = get_datasets(y,labels)
    options = set_title(title)
    
    chart = {
        'type': kind,
        'data': {
            'labels': x,
            'datasets': datasets
        },
    'options': options
    }
    
    return chart
```

```python
def get_api_chart(chart):
    url_base = 'https://quickchart.io/chart'
    resp = r.get(f'{url_base}?c={str(chart)}')
    return resp.content
```

```python
def save_image(path, content):
    with open(path, 'wb') as image:
        image.write(content)
```

```python
from PIL import Image
from IPython.display import display
```

```python
def display_image(path):
    img_pil = Image.open(path)
    display(img_pil)
```

```python
y_data_1 = []
for i in final_data[1::30]:
    y_data_1.append(i[0])
    
y_data_2 = []
for i in final_data[1::30]:
    y_data_2.append(i[1])
    
labels = ['Confirmados', 'Recuperados']

x = []
for i in final_data[1::30]:
    x.append(i[4].strftime('%d/%m/%Y'))
    
chart = create_chart(x, [y_data_1, y_data_2], labels, title='Confirmados x Recuperados - COVID-19')
chart_content = get_api_chart(chart)
save_image('dados/meu-primeiro-grafico.png', chart_content)
display_image('dados/meu-primeiro-grafico.png')
```

```python
from urllib.parse import quote
```

```python
def get_api_qrcode(link):
    text = quote(link)
    url_base = 'https://quickchart.io/qr'
    resp = r.get(f'{url_base}?text={text}')
    return resp.content
```

```python
url_base = 'https://quickchart.io/chart'
link = f'{url_base}?c={str(chart)}'
save_image('dados/qr-code.png', get_api_qrcode(link))
display_image('dados/qr-code.png')
```
