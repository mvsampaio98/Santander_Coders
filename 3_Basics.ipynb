**Manipulação de Arquivos**

```python
# Lendo o arquivo inteiro
arquivo = open('dados/noticia.txt', 'r', encoding='utf-8')
texto = arquivo.read()
print(texto)
arquivo.close()
```

```python
# Ler o arquivo linha a linha
arquivo = open('dados/noticia.txt', 'r', encoding='utf-8')
for linha in arquivo:
    print(linha, end='')
    linha = arquivo.readline()
arquivo.close()
```

```python
# Com o with não precisamos ficar fechando toda hora (.close())
with open('dados/noticia.txt', 'r', encoding='utf-8') as arquivo:
    texto = arquivo.read()
    print(texto)
```

```python
# Escrevendo com o Python no arquivo
with open('dados/noticia.txt', 'w', encoding='utf-8') as arquivo:
    arquivo.write('Essa é uma linha que eu escrevi usando Python\n')
    arquivo.write('Essa é a segunda linha que eu escrevi usando Python\n')
```

```python
# Vizualisando o que foi escrito
with open('dados/noticia.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read(), end='')
```

```python
# Também escreve
with open('dados/noticia.txt', 'a', encoding='utf-8') as arquivo:
    arquivo.write('Essa é a terceira linha que eu escrevi usando Python\n')
```

```python
# Vizualisando o que foi escrito
with open('dados/noticia.txt', 'r', encoding='utf-8') as arquivo:
    print(arquivo.read(), end='')
```

**Arquivos csv**

```python
import csv
```

```python
# Leitura de arquivo csv
with open('dados/ds_salaries.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    for linha in leitor:
        print(linha)
```

```python
# Vamos pegar um arquivo zerado para escrever nele
with open('dados/users.csv', 'w', encoding='utf-8', newline='') as arquivo_users:
    escritor = csv.writer(arquivo_users)
    escritor.writerow(['nome', 'sobrenome', 'email', 'genero'])
    escritor.writerow(['Marcelo','Sampaio','mvsampaio98@gmail.com','M'])
```

```python
# Fazendo um Script de cadastramento

head = ['nome', 'sobrenome']
dados = []
opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n')
while opt != '0':
    nome = input('Qual o seu nome? ')
    sobrenome = input('Qual o seu sobrenome? ')
    dados.append([nome, sobrenome])
    opt = input('O que deseja fazer?\n1 - Cadastrar\n0 - Sair\n')

with open('dados/cadastro.csv', 'w', encoding='utf-8', newline='') as arquivo_users:
    escritor = csv.writer(arquivo_users)
    escritor.writerow(head)
    escritor.writerows(dados)
    
with open('dados/cadastro.csv', 'r', encoding='utf-8') as arquivo_csv:
    leitor = csv.reader(arquivo_csv)
    for linha in leitor:
        print(linha)
```
